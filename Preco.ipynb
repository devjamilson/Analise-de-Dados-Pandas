```python
# pacotes usados neste notebook
import pandas as pd
import re
```

```python
#abrindo a base de dados
df = pd.read_csv('/content/drive/MyDrive/arquivo/2004-2021.csv')
```

```python
#Organizando e visualizando os dados

df = df.drop(columns=['DESVIO PADRÃO REVENDA'])
df = df.drop(columns=['PREÇO MÍNIMO REVENDA'])
df = df.drop(columns=['MARGEM MÉDIA REVENDA'])
df = df.drop(columns=['COEF DE VARIAÇÃO REVENDA'])
df = df.drop(columns=['PREÇO MÉDIO DISTRIBUIÇÃO'])
df = df.drop(columns=['DESVIO PADRÃO DISTRIBUIÇÃO'])
df = df.drop(columns=['PREÇO MÍNIMO DISTRIBUIÇÃO'])
df = df.drop(columns=['PREÇO MÁXIMO DISTRIBUIÇÃO'])
df = df.drop(columns=['COEF DE VARIAÇÃO DISTRIBUIÇÃO'])

df
```

```python
#exemplificando o que foi informado anteriormente
#obs - o nome da coluna pode conter espaço

jogadores_df = pd.DataFrame({
    'nome': ['Jose', 'Joao', 'Calo'],
    'idade': [16, 18, 17],
    'peso': [60.5, 75, 65],
})

jogadores_df

jogadores_df.info()
```
